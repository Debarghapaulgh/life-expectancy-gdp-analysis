### Life Expectancy Trends Over Time

To begin the analysis, we examine how life expectancy has changed over time across different countries. 

This visualization provides insight into long-term health trends and allows for comparison between countries in terms of improvement and overall levels of life expectancy.

By observing these patterns, we can better understand how health outcomes have evolved and identify differences in progress among countries.

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data = df,
    x = "Year",
    y = "Life expectancy at birth (years)",
    hue = "Country",
    marker = "o"
)
plt.title("Life Expectancy Trends Across Countries (2000–2015)", fontsize = 14)
plt.legend(title = "Country", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.tight_layout()
plt.show()
```

### Key Observations and Insights

- Life expectancy increases across all six countries from 2000 to 2015, indicating overall improvements in healthcare and living conditions.

- Zimbabwe shows the most dramatic change, with an initial decline followed by a strong recovery after 2004, rising from around 44 years to over 60 years.

- Germany consistently maintains the highest life expectancy, exceeding 80 years by the end of the period.

- Chile and the United States form a high-life-expectancy group, with Chile maintaining slightly higher values and showing stronger growth over time.

- China demonstrates steady and consistent improvement, gradually narrowing the gap with Mexico.

- Mexico shows overall growth with minor fluctuations during the mid-period.

- Despite overall progress, a significant gap remains between Zimbabwe and the other countries, highlighting persistent global inequalities in health outcomes.

### GDP Trends Over Time

To better understand the economic context of the countries in this dataset, we examine how Gross Domestic Product (GDP) changes over time.

This visualization highlights differences in economic growth patterns and allows for comparison of the scale and rate of economic development across countries between 2000 and 2015.

```python
df["GDP_trillion"] = df["GDP"] / 1e12
plt.figure(figsize = (10,6))
sns.lineplot(
    data = df,
    x = "Year",
    y = "GDP_trillion",
    hue = "Country",
    marker = "o"
)
plt.title("GDP Trends Across Countries (2000 - 2015)")
plt.ylabel("GDP( Trillion USD)")
plt.legend(title = "Country", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.tight_layout()
plt.show()
```

### Key Observations and Insights

- The large disparity in GDP causes the trends of Chile, Mexico, and Zimbabwe to appear compressed near the bottom of the graph, making detailed comparisons difficult on a linear scale.

- China exhibits the fastest GDP growth, overtaking Germany around 2007 and reaching the second-highest GDP by 2015.

- The United States maintains the highest GDP throughout the period while showing strong and generally consistent growth.

- Germany experiences steady economic growth during the first half of the period, followed by a plateau after 2008 and a slight decline towards 2015.

- A decline in GDP is visible for both the United States and Germany between 2008 and 2009, coinciding with the global financial crisis.

- Unlike the United States and Germany, China continues to grow throughout this period, although the rate of growth appears to slow temporarily.

### Relationship Between GDP and Life Expectancy

Economic growth is often associated with improvements in healthcare, infrastructure, nutrition, and overall quality of life. As a result, countries with stronger economies may be expected to exhibit higher life expectancy.

To investigate this relationship, a scatter plot is used to compare GDP and life expectancy across different countries and years. Since GDP values vary by several orders of magnitude, a logarithmic scale is applied to the GDP axis. This allows both smaller and larger economies to be represented more effectively and makes underlying patterns in the data easier to observe.

The visualization aims to identify whether higher levels of economic output are associated with longer life expectancy and to examine the strength of this relationship across the dataset.

```python
df["GDP_trillion"] = df["GDP"] / 1e12
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x="GDP_trillion",
    y="Life expectancy at birth (years)",
    hue="Country",
    s=100
)

plt.title("GDP vs Life Expectancy")
plt.xscale('log')
plt.xlabel("GDP in Trillion USD (log scale)")
plt.ylabel("Life Expectancy (years)")
plt.legend(title = "Country", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.tight_layout()
plt.show()
```

### Key Observations and Insights

- A clear positive relationship exists between GDP and life expectancy, with countries generally exhibiting higher life expectancy as GDP increases.

- The relationship is not perfectly linear. While economic growth is associated with improvements in life expectancy, the gains become smaller at higher levels of GDP, suggesting diminishing returns beyond a certain threshold.

- Chile emerges as a particularly interesting case. Despite operating at a significantly smaller economic scale than the United States and China, it achieves life expectancy levels comparable to or even higher than those countries. This suggests that factors beyond economic output, such as healthcare systems, public policy, and quality of life, play an important role in determining health outcomes.

- Germany, Chile, and the United States form a high-life-expectancy group, with all three countries reaching values close to or above 80 years despite substantial differences in economic size.

- China demonstrates the most pronounced economic expansion in the dataset. Its points span the largest horizontal distance on the graph, reflecting rapid GDP growth accompanied by steady improvements in life expectancy from the low 70s to the mid-70s.

- Zimbabwe forms a distinct cluster characterized by both the lowest GDP and the lowest life expectancy values. The country exhibits a dramatic recovery, with life expectancy rising from below 45 years to over 60 years alongside modest economic growth. This highlights the significant impact that improvements in economic and public health conditions can have on population well-being.

- The clustering of points by country indicates that GDP alone does not fully explain differences in life expectancy. Countries with similar life expectancy levels can operate at vastly different economic scales.

- Overall, the visualization suggests that economic development contributes positively to life expectancy, particularly for lower-income countries. However, beyond a certain level of economic prosperity, additional gains in GDP do not necessarily translate into proportionally higher gains in longevity.

## Correlation Analysis

While visualizations provide an intuitive understanding of the relationship between GDP and life expectancy, correlation analysis offers a quantitative measure of their association.

To better understand how strongly GDP and life expectancy move together over time, Pearson correlation coefficients are calculated separately for each country. Performing the analysis on a per-country basis avoids the distortions that can arise from large differences in economic scale between countries.

```python
corrs = []

for country in df["Country"].unique():
    country_df = df[df["Country"] == country]
    corr = country_df["GDP"].corr(country_df["Life expectancy at birth (years)"])
    corrs.append([country, round(corr, 2)])

corr_df = pd.DataFrame(corrs, columns=["Country", "Correlation"])
corr_df
```

### Key Observations and Insights

- All countries exhibit strong positive correlations between GDP and life expectancy, with correlation coefficients ranging from 0.91 to 0.98.

- The consistently high values indicate that periods of economic growth were generally accompanied by improvements in life expectancy.

- The United States and Zimbabwe show the strongest correlations in the dataset, suggesting a particularly close association between economic performance and life expectancy during the observed period.

- China, Germany, Mexico, and Chile also demonstrate strong positive relationships, reinforcing the overall trend observed in the visualizations.

- While these results support the existence of a positive association between GDP and life expectancy, correlation alone does not establish causation. Other factors such as healthcare access, education, public policy, and social conditions may also contribute significantly to improvements in life expectancy.

- Overall, the correlation analysis quantitatively supports the patterns observed in the visualizations, indicating that economic growth and life expectancy tended to improve together across all countries in the dataset.
